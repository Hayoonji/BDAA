### 행 단위 데이터 추출 시 꼭 알아야 하는 내용

- loc: 인덱스 기준으로 행 데이터 추출
- iloc: 행번호 기준으로 행 데이터 추출

```python
df=pd.read_csv('gapminder.tsv',sep='\t') #tab으로 구분되어 있음
```

### loc 이용해서 인덱스기준 데이터 추출할때 변수를 입력해서도 가능

```python
number_of_rows=mpg.shape[0] #mpg 데이터의 행개수 가져오기
```

```python
number_of_index=number_of_rows-12 #234-12
```

```python
number_of_index
```

```python
mpg
```

```python
mpg.loc[number_of_index] #인덱스번호가 222인 행 가져오기
```

```python
mpg.loc[[1,10,100]] #인덱스 번호 여러개 가져오기(대괄호 2개 반드시 사용)
```

### iloc?

```python
mpg
```

```python
mpg.loc[1] #인덱스 번호 기준
```

```python
df=mpg.loc[[1,10,100]]
```

```python
df.iloc[10] #인덱스번호로 추출 불가능(행이 3개밖에 없으니까)
```

```python
df.iloc[3]#인덱스번호로 추출 불가능(행이 3개밖에 없으니까)
```

```python
df.iloc[2] # 행번호로 추출가능(2번째 행이 있으니까)
```

```python
df=pd.read_csv('gapminder.tsv',sep='\t')
```

```python
df.groupby('year')['pop'].mean() #year별 'pop'의 평균값
```

```python
# year로 1차로 묶고 lifeExp별 pop과 gdpPercap의 평균값
df_gr=df.groupby(['year','lifeExp'])[['pop','gdpPercap']].mean() 
```

```python
#lifeExp로 1차로 묶고 year별 pop과 gdpPercap의 평균값
df.groupby(['lifeExp','year'])[['pop','gdpPercap']].mean()
```

```python
df_gr.info()
```

```python
df_gr
```
