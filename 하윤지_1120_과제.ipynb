### 필수과제

1. 스스로 mpg데이터를 통해 질문을 하고 그 질문에 답을 찾는 코드를 작성하기 , 주석처리 꼭 해주세요!
- ex) audi a4의 평균 연비는 몇일까?

- ex) midsize의 가장 연비가 좋은 차는 무엇일까?

- ex) model별로 가장 도시연비가 좋은 차는?  복합연비 (cty, hwy) 합의 평균 

- 너무 쉬운 질문의 답변 코드는 제외할게요. sum, mean, 기본적인 건 제외 최소 2개 이상 조건이 들어가야 함

질문 수는 최소 15개 이상으로 진행!

```python
import pandas as pd
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
# displ가 3 이상, 혹은 cyl가 6 이상인 행들을 출력해보자
mpg.query('displ>=3 | cyl>=6')
```

```python
# 2000년 전에 출시된 audi를 출력해보자
mpg.query('year<2000 & manufacturer == "audi"')
```

```python
# 도시연비가 20이하이면서 고속연비가 30이히인 행들의 평균 displ을 출력해보자
mpg.query('cty<=20 & hwy<=30')['displ'].mean()
```

```python
# 2000년 이전 출시되고 cyl가 4개 이하인 차량들의 도시연비의 평균 값을 출력해보자
mpg.query('year<2000 & cyl<=4')['cty'].mean()
```

```python
# 도시연비, 고속연비의 평균을 값으로 가지는 cty_hwy 열을 만들고 열 값에 따라 오름차순으로 정렬해보자
mpg['cty_hwy'] = (mpg['cty']+mpg['hwy']) / 2
mpg.sort_values('cty_hwy', ascending = True)
```

```python
# 제조사와 모델명별로 groupby를 통하여 cty_hwy의 평균을 출력해보자
mpg.groupby(['manufacturer','model'])['cty_hwy'].mean()
```

```python
# 연도와 카테고리별로 groupby를 통하여 개수를 출력해보자
mpg.groupby(['year','category'])['displ'].count()
```

```python
# 제조사명과 모델명을 합친 문자열을 detail이라는 열로 추가하자.
# 이후 detail별로 groupby를 통해 cty_hwy의 평균을 구하고 오름차순으로 정렬해보자
mpg.assign(detail = lambda x: x['manufacturer']+' '+x['model']).groupby('detail')['cty_hwy'].mean().sort_values(ascending = True)
```

```python
# 제조사별로 고속연비의 평균을 구하고 오름차순으로 정렬해보자
mpg.groupby('manufacturer').agg(mean_hwy = ('hwy','mean')).sort_values(by = 'mean_hwy', ascending = True)
```

```python
# 제조사별 groupby를 통하여 cyl의 평균값, cty_hwy의 중앙값, 차량의 개수를 구해보자
mpg.groupby('manufacturer').agg(cyl_mean = ('cyl','mean'),
                                cty_hwy_median = ('cty_hwy','median'),
                                n = ('manufacturer','count'))
```

```python
# 연도별 groupby를 통하여 도시연비의 평균값, 고속연비의 중앙값, 차량의 개수를 구해보자
mpg.groupby('year').agg(cty_mean = ('cty','mean'),
                        hwy_median = ('hwy','median'),
                        n = ('year','count'))
```

```python
# 모델명과 cty열을 추출하여 model_cty라는 변수에, 모델명과 hwy열을 추출하여 model_hwy변수에 저장하자
# 이후 모델명이 같은 행들끼리 합쳐서 model_cty_hwy로 병합해보자
model_cty = mpg[['model','cty']]
model_hwy = mpg[['model','hwy']]
model_cty_hwy = pd.merge(model_cty, model_hwy, how = 'inner')
model_cty_hwy
```

```python
# 2000년 이전에 생산되고 제조사명이 dodge와 ford인 차량들을 변수에 저장하고 병합해보자
dodge = mpg.query('year<2000 & manufacturer == "dodge"')
ford = mpg.query('year<2000 & manufacturer == "ford"')
dodge_ford = pd.concat([dodge, ford],axis=0)
dodge_ford
```

```python
# category의 종류가 minivan, suv인 차량들을 large 변수에, 나머지 차량들을 small 변수에 저장하고 도시연비, 고속연비의 평균값을 구해보자
large = mpg.query('category == "minivan" | category == "suv"')[['cty','hwy']].mean()
small = mpg.query('category != "minivan" & category != "suv"')[['cty','hwy']].mean()
```
