```python
import pandas as pd
mpg=pd.read_csv('mpg.csv')
df=pd.read_csv('gapminder.tsv',sep='\t')
```

```python
mpg
```

```python
# 1번: 제조사가 audi인 곳만
```

```python
audi=mpg['manufacturer']=='audi' #audi인 곳의 변수 지정
```

```python
mpg.loc[audi]
```

```python
# 2번 : 제조사가 hyndai 인곳인데 제조년이 2000년 이상인 것들
```

```python
mpg['manufacturer'].value_counts() # 제조사들의 이름들 확인
```

```python
hyundai_2000=(mpg['manufacturer']=='hyundai') & (mpg['year']>2000)#hyundai인 곳의 변수 지정
```

```python
df_hyundai_2000=mpg.loc[hyundai_2000] #loc에 변수 넣어서 df로 보여주기
df_hyundai_2000
```

```python
# 3번 : df_hyundai_2000의 끝에 5개 행만 가져오기
```

```python
df_hyundai_2000.iloc[-5:,:]
```

```python
# 4번 : df_hyundai_2000의 앞에 5개열 가져오기
```

```python
df_hyundai_2000.iloc[:,:5]
```

```python
# 5번 : 3,4번을 한꺼번에 처리(df_hyundai_2000의 끝 5개행과 앞의 5개열 가져오기)
```

```python
df_hyundai_2000.iloc[-5:,[0,1,2,3,4]]
```
